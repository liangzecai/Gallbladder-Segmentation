```python
from PIL import Image
import plotly.offline as py
import plotly.graph_objs as go
import pandas as pd
import numpy as np

py.init_notebook_mode(connected=True)
```

### RGB

```python
## curoff for binary mask
cutoff = 0.5

## load sample image and its mask
image = Image.open('new_image_04.jpg').resize((224,224))
mask = Image.open('mask4.png')
```

```python
## image where the Gb is segmented out
image_seg = np.multiply(np.array(image), np.array(mask) > cutoff)
```

```python
roi = Image.fromarray(image_seg.astype(np.uint8))
```

```python
roi
```

```python
def hg_trace(name, color, hg):
    line = go.Scatter(
        x=list(range(0,256)), 
        y=hg, 
        name=name,
        line=dict(color=(color)),
        mode='lines',
        showlegend=False
    )
    
    fill = go.Scatter(
        x=list(range(0,256)),
        y=hg,
        mode='none',
        name=name,
        line=dict(color=(color)), 
        hoverinfo='none'
    )
    
    return line, fill
```

```python
hg = roi.histogram()
# Returns a 768 member array with counts of R, G, B values
rhg = hg[0:256]
ghg = hg[256:512]
bhg = hg[512:768]

data = [*hg_trace('Red','red', rhg), 
        *hg_trace('Green', 'green', ghg),
        *hg_trace('Blue','blue', bhg)]

layout = go.Layout(
    title='RGB Histogram',
)

figure = go.Figure(data=data, layout=layout)

py.iplot(figure,
         filename='plot')
```
