```python
#### prepare/load data

# mask values: 0 or 255
img = image.load_img("../data/sample_data/train_image/20180503_ANIMAL1_0min.jpg", 
                     target_size=(224, 224)) # shape(244,244,3)
mask = image.load_img("../data/sample_data/train_mask/20180503_ANIMAL1_0min_ROI.png", 
                      target_size=(224, 224)) # shape(244,244,3)

# PIL image object to np.array
x = image.img_to_array(img) # shape(244,244,3)
y = image.img_to_array(mask).mean(axis=2)/255 # shape(244,244)
```

```python
def mask_overlay(image, mask, color=(0, 1, 0), alpha = 0.3, img_format = "image"):
    """
    Helper function to visualize mask on the top of the image
    Params:    
        image: np.array size(244,244,3) or Image object size(244,244,3)
        mask: np.array size(244,244) values{0,1} or Image object size(244,244,3) values{0,255}
        color: the color of the mask (choose RGB channel)
        img_format: "image" from .jpg or "array" from image.img_to_array
    Return:
        overlayed image (np.array)
    """ 
    if img_format == "image":
        mask = (mask * np.array(color)).astype(np.uint8)
        image = (image * np.array(1)).astype(np.uint8)
    else: # img_format == "array"
        color = color * 255
        mask = (np.dstack((mask, mask, mask)) * np.array(color)).astype(np.uint8)
        image = image.astype(np.uint8)
    
    weighted_sum = cv2.addWeighted(image, 1 - alpha, mask, alpha, 0.)
    img = image.copy()
    RGB_index = sum(color * np.array((0,1,2))) # indicator of which RGB channel R:0, G:1, B:2
    ind = mask[:, :, RGB_index] > 0    
    img[ind] = weighted_sum[ind]    
    return img
```

```python
### visualize image/mask
plt.imshow(mask_overlay(img, mask, img_format = "image"))
```

### loss function & metrics

```python
import keras.backend as K
from keras.losses import binary_crossentropy

def jaccard(y_true, y_pred):
    intersection = K.sum(K.abs(y_true * y_pred), axis=-1)
    union = K.sum(K.abs(y_true) + K.abs(y_pred), axis=-1) - intersection
    return (intersection + 1e-15) / (union + 1e-15)

def dice(y_true, y_pred):
    intersection = K.sum(K.abs(y_true * y_pred), axis=-1)
    union = K.sum(K.abs(y_true) + K.abs(y_pred), axis=-1)
    return (2 * intersection + 1e-15) / (union + 1e-15)


def customLoss(y_true, y_pred):
    H = binary_crossentropy(y_true, y_pred)
    J = jaccard(y_true, y_pred)
    return (H - K.log(J))
```

### train model

- binary segmentation

```python
print(x.shape)
print(y.shape)
```

```python
preprocessing_fn = get_preprocessing(BACKBONE)
x = preprocessing_fn(x)
```

```python
x = x.reshape((1,224,224,3))
y = y.reshape((1,224,224,1))
```

```python
# prepare model
model = Unet(backbone_name=BACKBONE, encoder_weights='imagenet')
model.compile('Adam', loss = 'binary_crossentropy', metrics = ['binary_accuracy',jaccard, dice])

# train model
history = model.fit(x, y, epochs=10,
                    verbose = 1)
```

```python
# model result
ypred = model.predict(x)
```
