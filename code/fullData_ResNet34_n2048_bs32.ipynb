### parameters

```python
SEED = 1234 ## for sample data, and data augmentation

### define backbone name
BACKBONE = 'resnet34'

### number of training data
n_train = 2048

### data path
x_train_path = "../data/image/image_train/"
y_train_path = "../data/mask/mask_train/"

x_test_path = "../data/image/image_validate/"
y_test_path = "../data/mask/mask_validate/"

### output path
outdir_path = "../result/full_model/" + BACKBONE + "_" + str(n_train) + "/"


### data augmentation parameters
data_gen_args = dict(#rescale=1./255,
                     rotation_range=30,
                     width_shift_range=0.1,
                     height_shift_range=0.1,
                     shear_range = 0.05,
                     zoom_range=0.1,
                     horizontal_flip=True,
                     fill_mode='nearest')

### model trainig
FIX_EPOCHS = 2
EPOCHS = 100
BATCH_SIZE = 32
```

### sample & load data

```python
### load train data
random.seed(SEED)
x_train_sample_list = random.sample(sorted(os.listdir(x_train_path)), n_train)
random.seed(SEED)
y_train_sample_list = random.sample(sorted(os.listdir(y_train_path)), n_train)

x_train = sample_image_from_folder(x_train_path, x_train_sample_list)
y_train = sample_image_from_folder(y_train_path, y_train_sample_list)
```

```python
## load test data
x_test =load_image_from_folder(x_test_path)
y_test =load_image_from_folder(y_test_path)
```

```python
### turn mask into non-RGB, binary
y_train = y_train.mean(axis=3)/255
y_test = y_test.mean(axis=3)/255

### reshape mask to tensor
y_train = y_train.reshape((-1, 224, 224,1))
y_test = y_test.reshape((-1, 224, 224,1))
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

- binary segmentation

### data augmentation

```python
### preprocessing based on BACKBONE
## for ResNet: RGB to BRG
preprocessing_fn = get_preprocessing(BACKBONE)
x_train = preprocessing_fn(x_train)
x_test = preprocessing_fn(x_test)
```

```python
image_datagen = image.ImageDataGenerator(**data_gen_args)
mask_datagen = image.ImageDataGenerator(**data_gen_args)

### no standardization, no fit
#image_datagen.fit(x_train, augment=True, seed=SEED)
#mask_datagen.fit(y_train, augment=True, seed=SEED)


image_generator = image_datagen.flow(x_train, batch_size=BATCH_SIZE, seed=SEED)
mask_generator = mask_datagen.flow(y_train, batch_size=BATCH_SIZE, seed=SEED)

# combine generators into one which yields image and masks
train_generator = zip(image_generator, mask_generator)
```

## train model

```python
# prepare model
model = Unet(backbone_name=BACKBONE, encoder_weights='imagenet', freeze_encoder=True)
model.compile('Adam', loss = customLoss, metrics = ['binary_accuracy', jaccard, dice])

# pretrain model decoder (freeze encode weights for 2 epochs)
model.fit_generator(train_generator, verbose = 0,
                    epochs = FIX_EPOCHS,
                    steps_per_epoch = np.floor(len(x_train)/(BATCH_SIZE)))

# release all layers for training
set_trainable(model)

# train model
history = model.fit_generator(
                    train_generator,
                    validation_data=(x_test,y_test),
                    epochs = EPOCHS,
                    steps_per_epoch = np.floor(len(x_train)/(BATCH_SIZE)),
                    verbose = 1)
```

```python
### save training log
RESULT_NAME = BACKBONE + "_n" + str(n_train) + "_batch" + str(BATCH_SIZE)
pd.DataFrame.from_dict(data=history.history).to_csv(outdir_path + RESULT_NAME + "_history.csv")
```

```python
# plot training log
for key in ['loss', 'binary_accuracy', 'jaccard', 'dice']:
    plot_history(history, key = key)
    plt.savefig(outdir_path + RESULT_NAME + "_model_" + key + ".jpg")
```

```python
model.save_weights(outdir_path + RESULT_NAME + ".h5")
```
